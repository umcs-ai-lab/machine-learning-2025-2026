## 1. Hello World + zmienne

**Wskazówka:** Wypisywanie: print('...'). Zmienna: name = 'Ala'.

```python
# Funkcja print() służy do wypisywania tekstu na ekran (w konsoli).
# Tekst (czyli tzw. string) zapisujemy w apostrofach '...' albo w cudzysłowie "..."
print('Hello, world!')  # Wypisujemy dokładnie tekst: Hello, world!


# Tworzymy zmienną o nazwie name.
# Zmienna to "pudełko", w którym przechowujemy dane.
# Po lewej stronie jest nazwa zmiennej, po prawej wartość.
# W tym przypadku przechowujemy tekst (string), czyli imię.
name = 'Jan'  # Można tu wpisać swoje imię jako tekst


# Chcemy wypisać zdanie: My name is <name>
# Możemy łączyć tekst ze zmienną na kilka sposobów.
# Tutaj używamy tzw. f-string (formatted string).
# Litera f przed tekstem pozwala wstawić zmienną w nawiasach klamrowych {}.
print(f'My name is {name}')  # Wypisujemy tekst wraz z wartością zmiennej
```

```python
# Poniższy blok sprawdza, czy zadanie zostało wykonane poprawnie.
try:
    # Sprawdzamy, czy zmienna name jest typu str (czyli czy jest tekstem).
    assert isinstance(name, str)

    # Jeśli wszystko jest poprawne, wypisujemy komunikat OK.
    print('OK ✅')

# Jeśli zmienna name w ogóle nie została utworzona,
# Python zgłosi błąd NameError.
except NameError:
    print('Brakuje zmiennej name')

# Jeśli zmienna istnieje, ale nie jest tekstem (np. liczbą),
# zgłoszony zostanie AssertionError.
except AssertionError:
    print('name powinno być typu str')
```

## 2. Liczby i proste obliczenia

**Wskazówka:** Dzielenie w Pythonie '/' zawsze daje float.


```python
# Tworzymy zmienną a i przypisujemy jej wartość 7 (liczba całkowita - int)
a = 7

# Tworzymy zmienną b i przypisujemy jej wartość 3 (również liczba całkowita - int)
b = 3


# Obliczamy sumę liczb a i b.
# Operator + oznacza dodawanie.
suma = a + b

# Obliczamy różnicę (odejmowanie).
# Operator - oznacza odejmowanie.
roznica = a - b

# Obliczamy iloczyn (mnożenie).
# Operator * oznacza mnożenie.
iloczyn = a * b

# Obliczamy iloraz (dzielenie).
# Operator / oznacza dzielenie.
# UWAGA: w Pythonie dzielenie zawsze zwraca liczbę typu float,
# nawet jeśli dzielimy dwie liczby całkowite.
iloraz = a / b
```

```python
try:
    # Sprawdzamy, czy a i b mają poprawne wartości.
    assert a == 7 and b == 3

    # Sprawdzamy, czy suma jest poprawna.
    assert suma == 10

    # Sprawdzamy różnicę.
    assert roznica == 4

    # Sprawdzamy iloczyn.
    assert iloczyn == 21

    # Sprawdzamy iloraz.
    # Używamy abs(), czyli wartości bezwzględnej.
    # 1e-12 oznacza bardzo małą liczbę: 0.000000000001
    # To zabezpieczenie przed błędami zaokrągleń w liczbach float.
    assert abs(iloraz - (7/3)) < 1e-12

    # Jeśli wszystko jest poprawne:
    print('OK ✅')

# Jeśli coś pójdzie nie tak:
except Exception as e:
    print('Coś nie gra:', e)
```

## 3. Stringi: łączenie i formatowanie

**Wskazówka:** Można użyć operatora + albo f-string: f'Hello, {full_name}!'


```python
# Mamy już dane wejściowe:
first = 'Jan'      # Zmienna first przechowuje imię (tekst)
last = 'Kowalski'  # Zmienna last przechowuje nazwisko (tekst)


# --- Budowanie full_name ---

# Łączymy teksty za pomocą operatora +
# Dodajemy między nimi spację ' '
# Uwaga: spacja też jest tekstem!
full_name = first + ' ' + last


# --- Budowanie greeting ---

# Używamy tzw. f-stringa (formatted string).
# Litera f przed tekstem pozwala wstawiać zmienne w nawiasach {}.
greeting = f'Hello, {full_name}!'
```

```python
try:
    assert full_name == 'Jan Kowalski'
    assert greeting == 'Hello, Jan Kowalski!'
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 4. Listy: tworzenie, indeksowanie, append

**Wskazówka:** Indeksy w Pythonie zaczynają się od 0.


```python
# Tworzymy listę o nazwie nums.
# Lista to uporządkowany zbiór elementów.
# Elementy zapisujemy w nawiasach kwadratowych [].
nums = [1, 2, 3]


# Metoda append() dodaje nowy element na koniec listy.
# W tym przypadku dodajemy liczbę 4.
nums.append(4)


# Indeksy w Pythonie zaczynają się od 0.
# To oznacza:
# indeks 0 → pierwszy element
# indeks 1 → drugi element
# indeks 2 → trzeci element
#
# Chcemy pobrać trzeci element, czyli ten o indeksie 2.
third = nums[2]
```

```python
try:
    assert nums == [1,2,3,4]
    assert third == 3
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 5. Pętla for: suma elementów

**Wskazówka:** W pętli: total += x


```python
# Tworzymy listę liczb.
nums = [2, 5, 1, 8]

# Tworzymy zmienną total i ustawiamy ją na 0.
# Będzie ona przechowywać sumę elementów.
total = 0


# Pętla for przechodzi po każdym elemencie listy nums.
# Zmienna x przyjmuje kolejno wartości: 2, potem 5, potem 1, potem 8.
for x in nums:

    # total +=x to skrócony zapis:
    # total = total + x
    # Czyli: dodaj aktualny element do sumy.
    total += x
```

```python
try:
    assert total == 16
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 6. if: policz ile liczb jest >= 5

**Wskazówka:** W pętli użyj if x >= 5.


```python
# Lista liczb
nums = [2, 5, 1, 8, 5, 3]

# Tworzymy licznik.
# Na początku jest 0, bo jeszcze nic nie policzyliśmy.
count_ge5 = 0


# Przechodzimy przez każdy element listy.
for x in nums:

    # Sprawdzamy warunek:
    # czy aktualna liczba jest większa lub równa 5?
    if x >= 5:

        # Jeśli warunek jest spełniony,
        # zwiększamy licznik o 1.
        count_ge5 += 1
```

```python
try:
    assert count_ge5 == 3
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 7. Mini-dane. Z tekstu do listy rekordów
W analizie danych często startuje się od surowego tekstu. Poniżej jest „mini-CSV” jako string.
Zrobimy z niego listę rekordów (listę słowników). Bez funkcji, krok po kroku.

```python
# Surowy tekst w formie "mini-CSV" (wielolinijkowy string).
csv_text = """name,age
Ala,20
Ola,22
Jan,19
"""

# Krok 1: usuwamy białe znaki z początku i końca (strip),
# a potem dzielimy tekst na linie po znaku nowej linii '\n'.
lines = csv_text.strip().split('\n')

# Krok 2: bierzemy pierwszą linię (nagłówek) i dzielimy ją po przecinku.
# Dostajemy listę nazw kolumn, np. ['name', 'age'].
header = lines[0].split(',')

# Krok 3: bierzemy wszystkie linie oprócz nagłówka.
# To są nasze wiersze z danymi.
rows = lines[1:]

# Tworzymy pustą listę, do której będziemy dodawać kolejne rekordy (słowniki).
people = []

# Przechodzimy pętlą po każdym wierszu z danymi (każda linia to jedna osoba).
for line in rows:
    # Dzielimy linię po przecinku.
    # Na przykład 'Ala,20' -> ['Ala', '20'].
    parts = line.split(',')

    # Pobieramy imię (pierwszy element listy).
    name = parts[0]

    # Pobieramy wiek (drugi element listy),
    # ale jest on jeszcze tekstem, więc zamieniamy go na int.
    age = int(parts[1])

    # Budujemy słownik (rekord) opisujący jedną osobę.
    # Klucze to nazwy pól, wartości to dane.
    person = {'name': name, 'age': age}

    # Dodajemy słownik do listy people.
    people.append(person)

# Wypisujemy wynik, żeby zobaczyć co powstało.
print('people =', people)
```

```python
# Sprawdzamy, czy wynik jest dokładnie taki jak oczekiwany.
try:
    assert people == [{'name':'Ala','age':20}, {'name':'Ola','age':22}, {'name':'Jan','age':19}]
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```
