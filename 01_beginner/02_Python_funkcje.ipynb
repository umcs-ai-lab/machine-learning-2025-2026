```python
## 1. Najprostsza funkcja
# Zobacz przykład i uruchom komórkę.

def add(a: int, b: int) -> int:
    # Funkcja przyjmuje dwa argumenty i zwraca ich sumę.
    return a + b

print(add(2, 3))  # 5
```

```python
## 2. Funkcja: zamień tekst na int (albo zwróć None)

# Wskazówka: Użyj try/except ValueError.

def safe_int(s: str) -> int | None:
    """
    Funkcja próbuje zamienić przekazany tekst na liczbę całkowitą.

    Parametry:
        s (str): tekst do konwersji

    Zwraca:
        int: jeżeli konwersja się powiedzie
        None: jeżeli tekst nie reprezentuje poprawnej liczby całkowitej
    """
    try:
        return int(s)
    except ValueError:
        return None

try:
    assert safe_int('10') == 10
    assert safe_int('-3') == -3
    assert safe_int('abc') is None
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

```python
## 3. Funkcja: proste czyszczenie tekstu (bez regex)

# Wskazówka:
# Najprościej: s.replace(',', ' ').replace('!', ' ') … potem ' '.join(s.split())

def clean_text(s: str) -> str:
    """
    Funkcja wykonuje proste czyszczenie tekstu:
    - zamienia litery na małe,
    - usuwa wybrane znaki interpunkcyjne przez zastąpienie ich spacją,
    - redukuje wielokrotne spacje do jednej,
    - usuwa spacje z początku i końca napisu.
    """

    # Zamiana wszystkich liter na małe.
    s = s.lower()

    # Zamiana przecinków i wykrzykników na spacje.
    s = s.replace(',', ' ').replace('!', ' ')

    # split() dzieli tekst po białych znakach i automatycznie usuwa ich nadmiar,
    # a join() skleja elementy pojedynczą spacją.
    s = ' '.join(s.split())

    return s

try:
    assert clean_text('  Hello,   WORLD!!!  ') == 'hello world'
    assert clean_text('Ala, ma kota!') == 'ala ma kota'
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

```python
## 4. Mini-pipeline: parsowanie jednej linii
# Zrobimy funkcję, która zamienia linię CSV na rekord (słownik).

line = 'Ala,20'

# Oczekiwany wynik:
# {'name': 'Ala', 'age': 20}
```

```python
## 4a. Funkcja parse_person

# Wskazówka: age = int(age_str)

def parse_person(line:str) -> str:
    """
    Funkcja przetwarza pojedynczą linię w formacie:
        imię,wiek

    i zwraca słownik:
        {'name': ..., 'age': ...}
    """

    # Rozbicie napisu na dwie części: imię i wiek zapisany jako tekst.
    name, age_str = line.split(',')

    # Dla większej odporności usuwamy ewentualne spacje wokół wartości.
    name = name.strip()
    age = int(age_str.strip())

    # Zwracamy dane w postaci słownika.
    return {'name': name, 'age': age}

try:
    assert parse_person('Ala,20') == {'name':'Ala','age':20}
    assert parse_person('Jan,19') == {'name':'Jan','age':19}
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```
